```python

from scipy.spatial.distance import mahalanobis
from scipy.stats import chi2
import scipy.stats as stats

df_fil=df[['price','minimum_nights', 'number_of_reviews','reviews_per_month', 'calculated_host_listings_count',
       'availability_365']]

mahal_distances = []
for row in df_fil.to_numpy():
  mahal_distances.append(mahalanobis(row, df_fil.mean(), np.linalg.inv(np.cov(df_fil.values.T))))
```
